```python
#Gets texture image from the sea ice data
def get_ice_image(D, mmin=None, mmax=None):
    #Normalise
    if mmin is not None and mmax is not None:
        image = (D-mmin)/(mmax-mmin)
        #Clip out of [0,1] range - in case defined range is not the global minima/maxima
        image = np.clip(image, 0, 1)
    else:
        image = D    
    return np.uint8(image*255)
```

```python
#TODO: Move these to accessvis!

#!pip install numpy-quaternion
#https://quaternion.readthedocs.io/en/latest/
import numpy as np
import quaternion as quat
import math

def magnitude(vec):
    return np.linalg.norm(vec)

def normalise(vec):
    norm = np.linalg.norm(vec)
    if norm == 0:
       vn = vec
    else:
        vn = vec / norm
    return vn

def rotate_vector(v1, v2):
    #Check for parallel or opposite
    v1 = np.array(v1)
    v2 = np.array(v2)
    if np.dot(v1, v2) > 0.999999:
        return quat.quaternion(1,0,0,0)
    elif np.dot(v1, v2) < -0.999999:
        #180 rotation about Y
        return quat.quaternion(1,0,1,0)
    xyz = np.cross(v1, v2)
    l1 = np.linalg.norm(v1)
    l2 = np.linalg.norm(v2)
    w = math.sqrt((l1*l1) * (l2*l2)) + np.dot(v1, v2)
    qr = quat.quaternion(w, xyz[0], xyz[1], xyz[2])
    return qr.normalized()

def set_camera(T, C=[0.,0.,-10]):
    # Set camera to follow above ship position
    # T = current ship coords
    
    #L = np.linalg.norm(T)
    #N = T / L
    #tr = N * L * 1.2 #1.2 * original vector length

    N = normalise(T)

    #q = vectorRotation([0,0,-1], N) #Align the model z axis
    q = rotate_vector(N, [0,0,1]) #Align the model z axis
    old_T = T

    #Now we have to correct alignment to view ship, apply angle adjustments
    #q2 = quat.from_euler_angles(np.radians(25), np.radians(-120), np.radians(-167))
    #q2 = quat.from_euler_angles(np.radians(0), np.radians(0), np.radians(0))
    #print(q2)
    #q = q * q2

    lv.rotation([q.x, q.y, q.z, q.w])
    #lv.rotation(25.726, 61.255, -167.494)

    #lv.rotation(25, 60, -167)
    #lv.rotate('z', 150)
    #lv.rotate('x', -30)

    #lv.rotate('z', 180)
    #lv.rotate('z', 75)
    lv.rotate('x', -60)
    lv.rotate('y', -10)
    lv.rotate('x', 5)

    #Even closer view
    lv.rotate('z', -5)
    lv.rotate('x', 5)

    #Final adjust for antarctic view
    lv.rotate('y', -25)
    lv.rotate('z', -15)
    lv.rotate('x', 10)

    lv.translation(C[0], C[1], C[2])

def align_model_vector(POS, heading):
    #POS is a sphere surface coord, so up vector / normal is same dir
    up = normalise(POS)
    
    #Look at rotation
    zdir = heading
    ydir = up
    xdir = np.cross(ydir, zdir)
    zdir = np.cross(xdir, ydir)
    rotm = np.array([xdir, ydir, zdir])
    rotm = rotm.T
    #print(rotm)
    #print(rotm.shape)
    q = quat.from_rotation_matrix(rotm) #, nonorthogonal=True)
    q = q.normalized()

    #Align the model axis
    q2 = quat.from_euler_angles(np.radians(90), np.radians(90), np.radians(270))

    q = q * q2
    
    #Return in LavaVu quaternion format
    return [q.x, q.y, q.z, q.w]
    #return q

#Just set to last point
print(route[0])
T0 = latlon_to_3D(route[0][1], route[0][2])
T1 = latlon_to_3D(route[1][1], route[1][2])
old_heading = T1 - T0
old_T = T0
print(old_heading, old_T)
#This function aligns the ship's heading vector given current point and saved previous point
def align_model(T, update_heading=True):
    global old_T, old_heading
    heading = T - old_T
    L = np.linalg.norm(heading)
    #heading = normalise(heading)
    if L > 0.0 and update_heading:
        old_T = T
        old_heading = heading
    else:
        heading = old_heading

    return align_model_vector(T, heading)
```

```python
def lerp(tr0, tr1, rot0, rot1, L):
    tr = tr0 + (tr1 - tr0) * L
    rot = rot0 + (rot1 - rot0) * L
    if len(rot) > 3:
        rot = rot / np.linalg.norm(rot) #Normalise quaternion
    lv.translation(*tr)
    lv.rotation(*rot)

def flight(tr0, tr1, rot0, rot1, steps, stop=False):
    lv.translation(*tr0)
    lv.rotation(*rot0)
    lv.render()

    for i in range(steps):
        if stop and i > stop:
            break
        L = i/(steps-1)
        lerp(tr0, tr1, rot0, rot1, L)
        lv.render()

def pause(frames=50):
    #Render pause
    for i in range(frames):
        lv.render()

def fly_in_1(steps=200):
    # Initial view (centred on prime meridian)
    tr0 = start_tr
    rot0 = start_rot
    
    # View of tierra del fuego and antarctica
    tr1 = cam_tr0
    rot1 = cam_rot0

    flight(tr0, tr1, rot0, rot1, steps)

def fly_in_2(stop=False):
    # Initial view (centred on prime meridian)
    #tr0 = np.array([0,0,-17])
    #rot0 = np.array([0,0,0,1])

    # Initial view (centred on Australia)
    tr0 = oz_tr0
    rot0 = oz_rot0

    # Europe zoomed view (Centered roughly on Antwerp)
    tr1 = route_tr0 #voyage_tr0
    rot1 = route_rot0 #voyage_rot0

    steps = 200
    flight(tr0, tr1, rot0, rot1, steps, stop)
```

```python
#Animate the voyage, defaults to full route data
def animate(start_date=route[0][0].date(), end_date=route[-1][0].date(), data_year=2023, day_steps=3, first=True, movecam=True):
    global overlay
    #Year for ice data
    tris0['visible'] = plot_ice

    pts.clear()
    lines_in.clear()
    lines_out.clear()
    lines_ice.clear()

    #start_date = date(1898, 1, 14) #Depart Staten Is for Antarctica
    #end_date = date(1899, 3, 28)   #Return to Punta Arenas

    delta = timedelta(days=1)
    pos = 0
    coord0 = None
    label = ''
    save_cam = None
    saved_ts = None
    tr0 = tr1 = rot0 = rot1 = None

    #Depart for antarctic
    antarctic = date(1898, 1, 29)

    #Reached antarctic
    antarctic = date(1898, 1, 29)
    #Show ship trapped between these dates
    trapped0 = date(1898, 3, 4)
    #trapped1 = date(1899, 2, 16) #Channel opens
    trapped1 = date(1899, 3, 14) #Open sea
    
    T1 = None
    update_heading = False
    if len(tris0.data) == 0:
        tris0.vertices(mesh) #Load static mesh rather than use timesteps

    cur_date = start_date
    while cur_date <= end_date:
        ts = cur_date

        if ts >= trapped0 and ts <= trapped1:
            snow = True
            update_heading = False
        else:
            snow  = False
            update_heading = True
    
        #Modify to data range by adding to year
        Y = ts.year+(data_year-1898)
        data_date = ts.replace(year=Y)
        D = get_ice_data_at(data_date)
        #if testrun:
        #    lv['title'] = str(ts) + ' [' + label + ']'

        #Plot the day of year / calendar overlay
        if ts >= date(1898, 1, 1):
            Im = plot_date(ts, data_date) #Show ice data date
        else:
            Im = plot_date(ts)
        #overlay.texture() #Clear texture
        overlay.texture(Im) #, flip=False)
        
        #Use previous years data if none available for current date?
        tris0.hide()
        if D is None:
            data_year -= 1
            Y = ts.year+(data_year-1898)
            data_date = ts.replace(year=Y)
            D = get_ice_data_at(data_date)
            
        #Have ice data for this date? Plot it
        if D is not None and plot_ice:
            raise_ice = False
            if raise_ice:
                #Multiply mesh vertices by D (raises thicker ice)
                V = mesh.copy()
                DZ = scipy.ndimage.zoom(D, [2,2], order=3)
                X = 1.0 + DZ*0.002
                Y = np.dstack((X,X,X))
                V = mesh * Y
                tris0.vertices(V)
                I = get_ice_image(DZ)
            else:
                #tris0.vertices(mesh)
                I = get_ice_image(D)
        
            #matplotlib.pyplot.imshow(I)
            tris0.texture(I, flip=False)
            #tris0.values(D) #.reshape((332, 316)))
            tris0.reload()
        #else:
        #    print('No data!')

        #Plot the route
        RV = None
        RP = None
        for i,r in enumerate(route):
            rdate = r[0].date()
            if ts == rdate:
                #Found the date
                RP = np.array([r[1], r[2]])
                #RV = np.array(r[1]), np.array(r[2]) #get_smoothed(r) #
                RV = latlon_to_3D(RP[0], RP[1])
                #print('FOUND',r[0],RV)
                #label = str(rdate)
                pts.vertices(RV)
                #Add description if present
                if len(r) > 3:
                    label = r[3] + ' - ' + str(rdate)
                    pts.labels(label)
                else:
                    pts.labels('')
                #print(ts, end=' ')
                pos = 0
                coord0 = np.array(r[1]), np.array(r[2])
                break
            elif rdate > ts:
                #No match: Interpolate a position between this and previous point
                pos += 1
                #Get days in range
                tdelta = rdate - route[i-1][0].date() #ts
                coord1 = np.array(r[1]), np.array(r[2])
                #Get lerp factor
                L = pos / tdelta.days #-1)
                #print(ts, rdate, tdelta.days, L)
                lat = coord0[0] + L * (coord1[0] - coord0[0])
                lon = coord0[1] + L * (coord1[1] - coord0[1])
                #RV = get_smoothed(r) #latlon_to_3D(np.array(lat), np.array(lon))
                RP = np.array([lat, lon])
                RV = latlon_to_3D(RP[0], RP[1])
                break

        #Render intermediate steps (per day of travel)
        POS = np.array([RV[0], RV[1], RV[2]])
        POS_LL = RP
        if T1 is not None:
            T00 = T0
            T0 = T1
            T1 = POS
            T00_LL = T0_LL
            T0_LL = T1_LL
            T1_LL = POS_LL

            if np.linalg.norm(T1-T0) == 0. and not snow:
                #Skip rendering days when there is no travel, except when stuck in ice
                cur_date += delta
                continue

            #Plot the previous day's route
            line_inc = 0.01 #3D coords
            #line_inc = 0.001 #lat,lon
            vec_inc = T0 - T00  #.copy() - T00.copy()
            ddist = np.linalg.norm(vec_inc)

            if not (T00==T0).all() and ddist > 0. and ddist < 1.:
                vec_inc = (vec_inc / ddist) * line_inc
                vert = T00.copy()
                for i in range(int(ddist//line_inc) + 1):
                    #vert_s = get_smoothed(vert)
                    if ts >= trapped0 and ts <= trapped1:
                        lines_ice.vertices(vert)
                    if ts < antarctic:
                        lines.vertices(vert)
                    if ts > antarctic and ts <= trapped0:
                        lines_in.vertices(vert)
                    if ts >= trapped1:
                        lines_out.vertices(vert)
                    vert += vec_inc
        else:
            T00 = T0 = T1 = POS
            T00_LL = T0_LL = T1_LL = POS_LL
            
        #r0 = get_rotation(T0)
        #T1 = POS #np.array([RV[0], RV[1], RV[2]])V =
        #r1 = get_rotation(T1)
        sub_steps = day_steps
        #If no movement, skip intermediate steps
        if (T0==T1).all():
            sub_steps = 1
        #When stuck in ice, reduce intermediate steps
        if snow and day_steps > 1:
            sub_steps = day_steps // 3 #2

        #Skip all camera movement and intra-day steps?
        if not movecam:
            #Use a fixed route point to aligns ship nicely during trapped sequence
            if not update_heading:
                r = route[99]
                RP = np.array([r[1], r[2]])
                RV = latlon_to_3D(RP[0], RP[1])
                POS0 = np.array([RV[0], RV[1], RV[2]])
                r = route[100]
                RP = np.array([r[1], r[2]])
                RV = latlon_to_3D(RP[0], RP[1])
                POS1 = np.array([RV[0], RV[1], RV[2]])
                heading = POS0-POS1
                transform(objlist, T0, align_model_vector(T0, heading), snow=snow)
            else:
                transform(objlist, T0, align_model(T0, update_heading), snow=snow)
            sub_steps = 0
            #TS = latlon_to_3D(np.array(T0[0]), np.array(T0[1]), np.array(0))
            #transform(objlist, T0, align_model(T0, update_heading), snow=snow)
            lv.render()
            #Render an additional frame for waves animation
            lv.render()

        for d in range(sub_steps):
            L = d/sub_steps
            T = T0 + (T1 - T0) * L
            T_LL = T0_LL + (T1_LL - T0_LL) * L
            #print(L,T)
            #rot = rot0 + (rot1 - rot0) * L
            #rot = rot / np.linalg.norm(rot) #Normalise quaternion
            #transform(objlist, T, get_rotation(T, heading_angle), snow=snow)

            #Use smoothed curve for ship heading alignment to avoid sudden direction changes
            if ts >= date(1898, 2, 22) and ts < trapped1:
                #Skip smoothing when trapped in ice
                if saved_ts is None:
                    saved_ts = get_smoothed(T0_LL)
                TS = saved_ts
            else:
                TS = get_smoothed(T_LL)
                saved_ts = TS
            transform(objlist, T, align_model(TS, update_heading), snow=snow)
            
            #Align camera with ship
            if ts <= antarctic:
                set_camera(T, C=INIT_CAM_TR)

            #Rotate around before trapped
            if ts >= antarctic: # and ts < trapped0:
                #Save cam position at this point
                if save_cam is None:
                    save_cam = lv.camera()
                    tr0 = np.array(save_cam['translate'])
                    rot0 = np.array(save_cam['rotate'])
                    #Set final adjusted rotation for trapped in ice section
                    #rot1 = np.array(cam['rotate'])
                    tr1 = [0.0, -3.905, -7.49]
                    rot1 = [-0.77, -0.107, 0.6, 0.188]
                    #Closer and higher view of sea ice
                    #tr1 = [0.0, -3.398, -8.652]
                    #rot1 = [-0.766, -0.138, 0.544, 0.314]
                    tr1 = [0.0, -3.707, -7.927]
                    rot1 = [-0.766, -0.126, 0.584, 0.239]

                #calc days until trapped
                if first and ts < trapped0:
                    dt = trapped0 - antarctic
                    dtc = trapped0 - cur_date
                    L = ((dt.days-dtc.days)*sub_steps + d)/(dt.days*sub_steps-1)
                else:
                    L = 1.0
                #print(L)
                lerp(tr0, tr1, rot0, rot1, L)
            
            lv.render()

        cur_date += delta
        #print(ts, end=' ')
        sys.stdout.flush()
        #break
    
    print('\nDone')
```
